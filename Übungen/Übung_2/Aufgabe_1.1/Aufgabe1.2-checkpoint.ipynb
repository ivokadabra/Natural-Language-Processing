```python
import nltk
import matplotlib.pyplot as pl
from nltk.probability import FreqDist
```

```python
file1 = open('tales/Aschenputtel.txt', 'r')
for line in file1:
    tokens_count=0;
    type_count=0;
    array_tokens=[]
    array_tokNum=[]
    array_typeNum=[]
    
    for word in nltk.word_tokenize(line):
    #array_tokens.append(word)
        tokens_count+=1
    #print(word+":"+ str(tokens_count))
        if (word in array_tokens):
            print(word+":"+ str(type_count)+" , "+str(tokens_count))
        else:
            type_count+=1
            print(word+":"+ str(type_count)+" , "+str(tokens_count))
            
        array_tokNum.append(tokens_count)
        array_typeNum.append(type_count)
        array_tokens.append(word) 
        
    pl.plot(array_tokNum,array_typeNum,'ro')
file1.close()        
```
