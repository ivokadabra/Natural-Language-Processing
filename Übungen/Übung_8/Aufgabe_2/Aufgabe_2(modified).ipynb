```python
def zeichen_entfernen(tokens):
    datensatz=[i.lower() for i in tokens if i.isalpha()]
    datensatz.append("</s>")
    datensatz.insert(0,"<s>")
    
    return datensatz
```

```python
def vizualisation(wc):

    #cancer = wc
    #print(cancer)
    
    res= [sub['val'] for sub in wc] 
    
    df = pd.DataFrame(res,columns=['d1','d2','d3','d4','d5'])
    features = ['d1','d2','d3','d4','d5']
    
    listofstrings=[sub['class'] for sub in wc]
    print(listofstrings)
#listofobjects = [classthing(s) for s in listofstrings]

    objs = {'val':listofstrings} 
    pf = pd.DataFrame(objs['val'],columns=['target'])
    pf.head()
    finalDf = pd.concat([df, pf['target']], axis = 1)
    finalDf.head()
    
    print(finalDf)
    features = ['d1','d2','d3','d4','d5','target']
    

    X = finalDf[['d1','d2','d3','d4','d5']]

    pca = PCA(n_components=3)
    components = pca.fit_transform(X)

    total_var = pca.explained_variance_ratio_.sum() * 100

    fig = px.scatter_3d(
       components, x=0, y=1, z=2, color=finalDf['target'],
       title=f'Total Explained Variance: {total_var:.2f}%',
       labels={'0': 'PC 1', '1': 'PC 2', '2': 'PC 3'}
     )
    fig.show()
    
    fig = px.scatter(components, x=0, y=1, color=finalDf['target'])
    fig.show()
```

```python
def tra(c_positive,wort,c_dict,c_positive_vec):
      for i in range(len(c_positive)):
            d={
                "val":c_positive_vec[i],
                 "word":c_positive[i] ,
                 "class":wort
            }
            c_dict.append(d)
       
      return c_dict
            
```
